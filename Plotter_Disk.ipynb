```python
import numpy as np
import scipy.stats as sc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_e_disk = np.genfromtxt("orbits_disk_e.csv", delimiter=",", dtype="float")
t = data_e_disk[:,101]
data_e_disk = np.delete(data_e_disk, [0,100], axis=1)
data_e_disk = np.delete(data_e_disk, [0,99], axis=1)
e_disk_mean = np.array([])
e_disk_std = np.array([])
for n in range(len(data_e_disk)):
    e_disk_std = np.append(e_disk_std, np.std(data_e_disk[n]))
    e_disk_mean = np.append(e_disk_mean,np.mean(data_e_disk[n]))
plt.figure(figsize=(20,10))
plt.plot(t,e_disk_mean, 'r-')
plt.plot(t,e_disk_mean + e_disk_std, 'b--')
plt.plot(t,e_disk_mean - e_disk_std,'b--')
plt.xlim(0,max(t))
plt.title("Eccentricity Over Time", fontsize = 18)
plt.xlabel("Orbits", fontsize = 16)
plt.ylabel("Eccentricity", fontsize = 16)
plt.fill_between(t, e_disk_mean + e_disk_std, e_disk_mean - e_disk_std, color='blue', alpha='0.5')
plt.savefig('EccentricityOverTime_1.png')
```
